```python
import pandas as pd
employment_df = pd.read_csv('Employment.csv')
pd.options.display.float_format = '{:,.0f}'.format
employment_df
```

```python
employment_df_industry = employment_df.groupby('Industry')
```

```python
industry_array=[]
for index, row in employment_df.iterrows():
    industry_array.append(employment_df.loc[index]["Industry"])
    
industry_array_final = []
for i in industry_array:
    if i not in industry_array_final:
        industry_array_final.append(i)
print(industry_array_final)
```

```python
import matplotlib.pyplot as plt
dataframes_arr=[]
x_all = []
y_all = []
for i in industry_array_final:
    
    df_industry1=employment_df_industry.get_group(i)
    
    for index, row in df_industry1.iterrows():
        if df_industry1.loc[index]["Year"]==2022:
            df_industry1.at[index,"Average"]=(float(df_industry1.loc[index]["Jan"])+float(df_industry1.loc[index]["Feb"])+float(df_industry1.loc[index]["Mar"]))/3
        else:
            df_industry1.at[index,"Average"]=(float(df_industry1.loc[index]["Jan"])+float(df_industry1.loc[index]["Feb"])+float(df_industry1.loc[index]["Mar"])+float(df_industry1.loc[index]["Apr"])+float(df_industry1.loc[index]["May"])+float(df_industry1.loc[index]["Jun"])+float(df_industry1.loc[index]["Jul"])+float(df_industry1.loc[index]["Aug"])+float(df_industry1.loc[index]["Sep"])+float(df_industry1.loc[index]["Oct"])+float(df_industry1.loc[index]["Nov"])+float(df_industry1.loc[index]["Dec"]))/12

#     dataframes_arr.append(df_industry1)
#     print(df_industry1)
    x_array=[]
    y_array=[]
    for index, row in df_industry1.iterrows():
        x_array.append(df_industry1.loc[index]["Year"])
        y_array.append(df_industry1.loc[index]["Average"])

    x_all.append(x_array)
    y_all.append(y_array)
    plt.plot(x_array, y_array)
    plt.xticks(rotation = 90)
    plt.xlabel('Year')
    plt.ylabel('Employment')
    plt.title (df_industry1.loc[index]["Industry"])
#     plt.legend(loc="upper right")
#     plt.ylim([0, 200])
    plt.show()
    


        
```

```python
fig, ax = plt.subplots(figsize=(15,10))
for i in range(len(y_all)):
    plt.plot(x_all[i], y_all[i])
# plt.plot(combined['All'].loc[2010:2021],color='green', label='All')
plt.title('Employment Data')
plt.xlabel('Year')
plt.legend()
plt.ylabel('')
```
