```python
from datetime import datetime
import matplotlib.pyplot as plt
unique_age_groups=["18 - 29 years","30 - 39 years","40 - 49 years","50 - 59 years","60 - 69 years","70 - 79 years","80 years and above"]
for age_g in unique_age_groups:
#     df_age1=df_age.get_group("18 - 29 years")
    df_age1=df_age.get_group(age_g)
    df_age1 = df_age1.reset_index(drop=True)
#     print(df_age1)
    unique_labels=df_age1["Time Period Label"].unique()
    values=dict.fromkeys(unique_labels,0)
#     date_format = "%m/%d/%Y"
    time_range=[]
    value_arr=[]
    for index, row in df_age1.iterrows():
#         start = datetime.strptime(df.iloc[index]["Time Period Start Date"], date_format)
#         end =   datetime.strptime(df.iloc[index]["Time Period End Date"], date_format)
#         diff = (end.date() - start.date()).days
    #     print(diff)
#         s=df.iloc[index]
    #     print(df.iloc[index]["Time Period Label"])

        values[df_age1.iloc[index]["Time Period Label"]]= values[df_age1.iloc[index]["Time Period Label"]]+df_age1.iloc[index]["Value"]

    for key in values:
        time_range.append(key)
        value_arr.append(values[key])
#     print(value_arr)
    plt.plot(time_range, value_arr)  # Plot the chart
    xmin = min(summ_arr) 
    xmax=max(summ_arr)
    for i, x in enumerate(summ_arr):
        summ_arr[i] = (x-xmin) / (xmax-xmin)
        summ_arr[i]*=150
    plt.plot(time_range, summ_arr)
    plt.xticks(rotation = 90)
#     plt.ylim([0, 200])
    plt.show()
```

```python
import pandas as pd
df_cases = pd.read_csv('us.csv')
df_cases
```

```python
for index, row in df_cases.iterrows():
    if index==0:
        df_cases.loc[index,"new_cases"]=1.0
        continue 
    df_cases.loc[index,"new_cases"]=df_cases.loc[index,"cases"]-df_cases.loc[index-1,"cases"]
    
```

```python
unique_labels_start=df["Time Period Start Date"].unique()
unique_labels_end=df["Time Period End Date"].unique()
```

```python
ans_labels_start=[]
for i in unique_labels_start:
    x=datetime.strptime(i,'%m/%d/%Y')
    newformat = x.strftime('%Y-%m-%d')
    
    print(newformat)
    ans_labels_start.append(newformat)
    
    
ans_labels_end=[]
for i in unique_labels_end:
    x=datetime.strptime(i,'%m/%d/%Y')
    newformat = x.strftime('%Y-%m-%d')
    
    print(newformat)
    ans_labels_end.append(newformat)
#     i.strftime('%m-%d-%Y')
```

```python
df_cases
```

```python
summ_arr=[]
for i in range(len(ans_labels_end)):
    summ=0
    for index, row in df_cases.iterrows():
        if df_cases.loc[index,'date']>=ans_labels_start[i] and df_cases.loc[index,'date']<=ans_labels_end[i]:
#             print(df_cases.loc[index,'date'])
#     print("NEWWWWWW")
            summ+=df_cases.loc[index]["new_cases"]
    print(summ)
    summ_arr.append(summ)
```

```python
print(len(time_range),len(summ_arr))
plt.plot(time_range, summ_arr)  # Plot the chart
plt.xticks(rotation = 90)
plt.show()
```
